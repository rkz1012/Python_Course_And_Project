```python
class BaseSalary:
  def __init__(obj,base_salary, bonus_rate = 0.1, symbol = '$'):
    obj.base_salary = base_salary
    obj.bonus_rate = bonus_rate
    obj.symbol = symbol
    obj.total_salary= (base_salary) * (1+bonus_rate)
    obj.bonus = (obj.total_salary - base_salary)

  def __repr__(obj):
     return f'{obj.symbol} {obj.base_salary:,.0f}'

  def calculate_salary(obj):
    return f'{obj.symbol} {obj.total_salary:,.0f}'

  def calculate_bonus(obj):
    return f'{obj.symbol} {obj.bonus:,.0f}'

```

```python
BaseSalary(500000)
```

```python
sal= BaseSalary(500000)

sal.base_salary
```

```python
sal.calculate_salary()
```

```python
sal.calculate_bonus()
```

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country'].value_counts().head(20)
```

```python
india_jobs = df[df['job_country' ] == 'India']
```

```python
india_jobs
```

```python
india_jobs = india_jobs[india_jobs['salary_year_avg'].notna()]

india_jobs.groupby('job_title_short')['salary_year_avg'].agg(['min','max','median','mean']).sort_values(by = 'median', ascending= False)
```

```python
# Use a dictionary to specify the new column names and aggregation functions
result = india_jobs.groupby('job_title_short')['salary_year_avg'].agg(
    min_salary='min',
    max_salary='max',
    median_salary='median',
    average_salary='mean'
)

print(result)
```
